### 1. Detect lines in the image
Use the Canny edge detector and Hough transform to detect lines in the image.

```python
import os
import numpy as np
import cv2  
from matplotlib import pyplot as plt


def canny_edge(img):
    rootpath='./COMP90086_asst3_2021_images/'
    gray = cv2.imread(os.path.join(rootpath,img),cv2.IMREAD_GRAYSCALE)
    gray = cv2.GaussianBlur(gray, (5, 5), 0)
    edge_img = cv2.Canny(gray,100,200,apertureSize=3,L2gradient=True)
    Lines_P = cv2.HoughLinesP(edge_img, 1, np.pi/180,30,minLineLength=50, maxLineGap=5)
    
    color_img_P = cv2.cvtColor(gray, cv2.COLOR_GRAY2RGB) 
    
    
    line_list = []
    #i=0
    for line in Lines_P:
        x1,y1, x2, y2 = line[0]
        angle = np.arctan2(y2 - y1, x2 - x1)* 180.0 /np.pi
        if 78>angle>12 or -12>angle>-78:
            #print(angle)
            #print(x1,y1,x2,y2)
            line_list.append([x1,y1,x2,y2])
            cv2.line(color_img_P, (x1, y1), (x2, y2), (255,0,0), 2)
            
        else:
            continue
    
    plt.subplots(figsize=(15, 15))
    plt.subplot(1,3,1)
    plt.imshow(gray, cmap='gray')  
    plt.title('Original image')
    plt.axis('off')

    plt.subplot(1,3,2)
    plt.imshow(edge_img, cmap='gray')  
    plt.title('Canny edge map')
    plt.axis('off')
    
    plt.subplot(1,3,3)
    plt.imshow(color_img_P)  
    plt.title('Detected Lines (in red)')
    plt.axis('off')
    plt.show()
    
    
    return line_list
```

### 2. Locate the vanishing point
Use RANSAC to locate the vanishing point from the detected lines.

#### 2.1 RANSAC functions
Define two fuctions required by RANSAC: a function to find the point where lines intersect, and a function to compute the distance from a point to a line.

```python
# Find the intersection point
import sys
import math
def find_intersection_point(line1,line2):
    x1, y1, x2, y2 = line1
    x3, y3, x4, y4 = line2
    
    ##a1x + b1y = c1 for line1
    a1 = (y2-y1)* 1.0
    b1 = (x1-x2)*1.0
    c1 = a1*x1+b1*y1;
    
    ####a2x +b2y = c2 for line 2
    a2 = (y4-y3)*1.0
    b2 = (x3-x4)*1.0
    c2 = a2*x3+b2*y3
    
    determinant = a1*b2 - a2*b1;
    
    if determinant ==0:
        return (float("inf"),float("inf"))
    else:
        point_x = (b2*c1 - b1*c2)/determinant;
        point_y = (a1*c2 - a2*c1)/determinant;
        
        #print(point_x,point_y)
        return (point_x,point_y)
        
        
# Find the distance from a point to a line
def find_dist_to_line(line,point):
    
    ### coed reference from week 7 worksheet
    x1, y1, x2, y2 = line
    m = (y2 - y1) / (x2 - x1 + sys.float_info.epsilon)
    c = y2 - m * x2
    
    if point == (float("inf"),float("inf")):
        return float("inf")
    else:
        (x0,y0)=point
        x = (x0 + m*y0 - m*c)/(1 + m**2)
        y = (m*x0 + (m**2)*y0 - (m**2)*c)/(1 + m**2) + c
        distance = math.sqrt((x - x0)**2 + (y - y0)**2)
        return distance
        

#     x1, y1, x2, y2 = line
#     (point_x,point_y)=point
#     a = (y2-y1)* 1.0
#     b = (x1-x2)*1.0
#     c = a*x1+b*y1;
#     distance = np.abs(a * point_x + b * point_y + c) / (np.sqrt(a**2 + b**2))
#     print("distance ",distance)
#     return distance 
```

#### 2.2 RANSAC loop
Define the main RANSAC loop

```python
import math 
import random
def ransac(lines,itera,thres,ration):
    
    for it in range(itera):
        num = 0
        random = np.random.randint(len(lines),size=2)
        point = find_intersection_point(lines[random[0]],lines[random[1]])
#         (x,y) = point
#         print("aaaa",point[0])
#         print("bb",point[1])
        if point!=None and point[0]<0 or point[1]<0:
            #print("point out of imange")
            continue
            
        for line in lines:
             if find_dist_to_line(line, point)<thres:
                    num = num+1
                    
        if num/len(lines)*1.0 >= ration:
            #print("The model is found !")
            #print("current reall ration is",num/len(lines)*1.0 )
            return point
#         else:
#             print("current in iterations ", it," ration is ",num/len(lines)*1.0)
                  
                  
    
    print("the model is not find")
    return None
    

                  
```
